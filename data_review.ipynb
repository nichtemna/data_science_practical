```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
```

```python
DF = pd.read_excel('data/data8.xls')
```

```python
# Percentage of rented properties in all occupied houses
DF['pctrnths'] = DF['rnthsg'] / DF['tothsg']
```

```python
len(DF)
```

```python
DF.columns.values.tolist()
```

```python
# Sort data by city population. With City 47 as much bigger than any other, that's the one that we see as outlier in many plots.
DF.sort_values(by=['pop'], ascending=False)
```

```python
print(DF[['city', 'year', 'pop', 'enroll', 'rent', 'avginc', 'pctrnths']])
```

```python
DF80 = DF[DF['year'] == 80]
DF90 = DF[DF['year'] == 90]

# Remove outlier city 47
DF80no = DF80[DF80['city'] != 47]
DF90no = DF90[DF90['city'] != 47]
DFno = DF[DF['city'] != 47]
```

```python
# Avereage income vs Average rent scatterplots
_, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 6))
ax1.scatter(DF80['avginc'], DF80['rent'])
ax2.scatter(DF90['avginc'], DF90['rent'])
ax3.scatter(DF80['lavginc'], DF80['lrent'])
ax4.scatter(DF90['lavginc'], DF90['lrent'])

ax1.set_title('1980')
ax1.set_xlabel("Income")
ax1.set_ylabel("Rent")

ax2.set_title('1990')
ax2.set_xlabel("Income")
ax2.set_ylabel("Rent")

ax3.set_title('1980')
ax3.set_xlabel("Log Income")
ax3.set_ylabel("Log Rent")

ax4.set_title('1990')
ax4.set_xlabel("Log Income")
ax4.set_ylabel("Log Rent")

plt.suptitle('Average income per capita vs Average rent')
plt.tight_layout()
plt.show()
```

```python
# Change in log values vs log average rent
_, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 6))
ax1.scatter(DF90['clavginc'], DF90['clrent'])
ax2.scatter(DF90['clpop'], DF90['clrent'])
ax3.scatter(DF90['cpctstu'], DF90['clrent'])
ax4.scatter(DF90['clrnthsg'], DF90['clrent'])

ax1.set_title('Change Log Income vs Change Log Rent')
ax1.set_xlabel("Change Log Income")
ax1.set_ylabel("Change Log Rent")

ax2.set_title('Change Log Population vs Change Log Rent')
ax2.set_xlabel("Change Log Population")
ax2.set_ylabel("Change Log Rent")

ax3.set_title('Change percent of population students vs Change Log Rent')
ax3.set_xlabel("Change percent of population students")
ax3.set_ylabel("Change Log Rent")

ax4.set_title('Change log renter occupied houses vs Change Log Rent')
ax4.set_xlabel("Change log renter occupied houses")
ax4.set_ylabel("Change Log Rent")

plt.suptitle('Change of log values vs Change of log rent')
plt.tight_layout()
plt.show()
```

```python
# Population vs Average rent scatterplots
_, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 6))
ax1.scatter(DF80['pop'], DF80['rent'])
ax2.scatter(DF90['pop'], DF90['rent'])
ax3.scatter(DF80['lpop'], DF80['lrent'])
ax4.scatter(DF90['lpop'], DF90['lrent'])

ax1.set_title('1980')
ax1.set_xlabel("Population")
ax1.set_ylabel("Rent")

ax2.set_title('1990')
ax2.set_xlabel("Population")
ax2.set_ylabel("Rent")

ax3.set_title('1980')
ax3.set_xlabel("Log Population")
ax3.set_ylabel("Log Rent")

ax4.set_title('1990')
ax4.set_xlabel("Log Population")
ax4.set_ylabel("Log Rent")

plt.suptitle('Population vs Average Rent')
plt.tight_layout()
plt.show()
```

```python
# Percent of students vs Average rent scatterplots
_, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.scatter(DF80['pctstu'], DF80['rent'])
ax2.scatter(DF90['pctstu'], DF90['rent'])

ax1.set_title('1980')
ax1.set_xlabel("Percent of students")
ax1.set_ylabel("Rent")

ax2.set_title('1990')
ax2.set_xlabel("Percent of students")
ax2.set_ylabel("Rent")

plt.suptitle('Percent of students vs Average Rent')
plt.tight_layout()
plt.show()
```

```python
# Number of occuppied houses vs rented houses
_, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 6))
ax1.scatter(DF80['tothsg'], DF80['rnthsg'])
ax2.scatter(DF90['tothsg'], DF90['rnthsg'])
ax3.scatter(DF80['ltothsg'], DF80['lrnthsg'])
ax4.scatter(DF90['ltothsg'], DF90['lrnthsg'])

ax1.set_title('1980')
ax1.set_xlabel("Houses")
ax1.set_ylabel("Rented houses")

ax2.set_title('1990')
ax2.set_xlabel("Houses")
ax2.set_ylabel("Rented houses")

ax3.set_title('1980')
ax3.set_xlabel("Log Houses")
ax3.set_ylabel("Log Rented houses")

ax4.set_title('1990')
ax4.set_xlabel("Log Houses")
ax4.set_ylabel("Log Rented houses")

plt.suptitle('Number of houses vs number of rented houses')
plt.tight_layout()
plt.show()
```

```python
# Percent of rented houses vs percent of students
_, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.scatter(DF80['pctrnths'], DF80['pctstu'])
ax2.scatter(DF90['pctrnths'], DF90['pctstu'])

ax1.set_title('1980')
ax1.set_xlabel("Percent of rented properties")
ax1.set_ylabel("Percent of students")

ax2.set_title('1990')
ax2.set_xlabel("Percent of rented properties")
ax2.set_ylabel("Percent of students")

plt.suptitle('Percent of rented to Percent of students')
plt.tight_layout()
plt.show()

# Seems like percent of rented properties is correlated with percent of students in population
```

```python
# Population
_, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(12, 8))

ax1.hist(DF80['pop'], bins=20)
ax2.hist(DF90['pop'], bins=20)
ax3.boxplot(DF.pivot(columns='year', values='pop', index=['city']), labels=['1980', '1990'])
ax4.hist(DF80no['pop'], bins=20)
ax5.hist(DF90no['pop'], bins=20)
ax6.boxplot(DFno.pivot(columns='year', values='pop', index=['city']), labels=['1980', '1990'])

ax1.set_title('1980')
ax1.tick_params(axis='x', rotation=90)
ax2.set_title('1990')
ax2.tick_params(axis='x', rotation=90)
ax3.set_title('Boxplot')
ax3.grid(True)
ax4.set_title('1980 no outlier')
ax4.tick_params(axis='x', rotation=90)
ax5.set_title('1990 no outlier')
ax6.tick_params(axis='x', rotation=90)
ax6.set_title('Boxplot no outlier')
ax6.grid(True)

plt.suptitle('Population')
plt.tight_layout()
plt.show()

# Population in cities has grown slightly.
# There is one outlier - city 47, which is much bigger than any other city.
# What if we remove it?
```

```python
# Average rent
_, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(12, 8))

ax1.hist(DF80['rent'], bins=20)
ax2.hist(DF90['rent'], bins=20)
ax3.boxplot(DF.pivot(columns='year', values='rent', index=['city']), labels=['1980', '1990'])
ax4.hist(DF80no['rent'], bins=20)
ax5.hist(DF90no['rent'], bins=20)
ax6.boxplot(DFno.pivot(columns='year', values='rent', index=['city']), labels=['1980', '1990'])

ax1.set_title('1980')
ax2.set_title('1990')
ax3.set_title('Boxplot')
ax3.grid(True)
ax4.set_title('1980 no outlier')
ax4.tick_params(axis='x', rotation=90)
ax5.set_title('1990 no outlier')
ax6.tick_params(axis='x', rotation=90)
ax6.set_title('Boxplot no outlier')
ax6.grid(True)


plt.suptitle('Average rent')
plt.tight_layout()
plt.show()

# - seems like average rent has grown significantly, median has grown almost twice.
# - removing outlier city 47 doesn't change the picture much
```

```python
# Average income
_, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))

ax1.hist(DF80['avginc'], bins=20)
ax2.hist(DF90['avginc'], bins=20)
ax3.boxplot(DF.pivot(columns='year', values='avginc', index=['city']), labels=['1980', '1990'])

ax1.set_title('1980')
ax2.set_title('1990')
ax3.set_title('Boxplot')
ax3.grid(True)

plt.suptitle('Average income per capita')
plt.tight_layout()
plt.show()

# - seems like average income has grown as well
```

```python
# Percent of population students
_, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))


ax1.hist(DF80['pctstu'], bins=20)
ax2.hist(DF90['pctstu'], bins=20)
ax3.boxplot(DF.pivot(columns='year', values='pctstu', index=['city']), labels=['1980', '1990'])

ax1.set_title('1980')
ax2.set_title('1980')
ax3.set_title('Boxplot')
ax3.grid(True)

plt.suptitle('Percent of population students')
plt.tight_layout()
plt.show()

# - seems like percent of students has grown a little bit
```

## Correlation analysis

```python
cdf = DF.corr(numeric_only=True)[['pop', 'pctstu', 'avginc', 'enroll', 'tothsg', 'rnthsg', 'pctrnths']]
cdf = cdf[cdf.index.isin(['rent', 'tothsg', 'rnthsg', 'pop', 'enroll', 'pctrnths'])]
cdf
```

- average rent is closely correlated with avginc (0.88) and number of enrolled college students (0.2)
- number of rented houses is correlated with total number of houses (0.62) and number of enrolled students (0.76)
- percent of rented houses is correlated with percent of students in population (0.6)
