```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
```

```python
DF = pd.read_excel('data/data8.xls')
```

```python
len(DF)
```

```python
DF.columns.values.tolist()
```

```python
DF.sort_values(by=['pop'])
```

```python
print(DF[['city', 'year', 'pop', 'enroll', 'rent', 'avginc']])
```

```python
DF.plot.scatter(x='pop', y='rent')
```

```python
DF80 = DF[DF['year'] == 80]
DF90 = DF[DF['year'] == 90]
```

```python
DF80.plot.scatter(x='pop', y='rent')
```

```python
DF90.plot.scatter(x='pop', y='rent')
```

```python
DF80.plot.scatter(x='lpop', y='lrent')
```

```python
DF90.plot.scatter(x='lpop', y='lrent')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.hist(DF80['pop'], bins=10)
ax2.hist(DF90['pop'], bins=10)
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.hist(DF80['enroll'], bins=10)
ax2.hist(DF90['enroll'], bins=10)
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.hist(DF80['rent'], bins=10)
ax2.hist(DF90['rent'], bins=10)
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.hist(DF80['avginc'], bins=10)
ax2.hist(DF90['avginc'], bins=10)
plt.tight_layout()
plt.show()
```
